## 5. Missing Value Assessment

Missing values can impact analysis, feature engineering, and model performance.

This step verifies whether any columns contain missing values and quantifies their extent.

```python
missing_values = df.isnull().sum()

missing_summary = pd.DataFrame({
    "Missing Values": missing_values,
    "Missing Percentage": (missing_values / len(df)) * 100
})

missing_columns = missing_summary[
    missing_summary["Missing Values"] > 0
]

print("Columns with missing values:", len(missing_columns))

missing_columns
```

### Findings

- No missing values were detected in the dataset.
- All 35 columns contain complete records.
- No imputation or missing value treatment is required.

## 6. Duplicate Record Assessment

Duplicate employee records can distort analysis results and model training.

This step verifies whether duplicate rows exist in the dataset.

```python
duplicate_count = df.duplicated().sum()

print("Duplicate Records:", duplicate_count)
```

### Findings

- No duplicate employee records were identified.
- The dataset contains 1,470 unique employee records.
- Duplicate removal is not required.

## 8. Constant Value Assessment

Columns containing only a single value do not provide analytical or predictive value.

This step identifies features with no variability.

```python
constant_columns = []

for col in df.columns:
    if df[col].nunique() == 1:
        constant_columns.append(col)

print("Constant Columns:")
print(constant_columns)
```

### Findings

Three constant-value columns were identified:

- EmployeeCount
- Over18
- StandardHours

These columns contain no variability across records and therefore do not contribute any predictive or analytical value.

They will be removed in the feature engineering phase to improve model efficiency and reduce noise.
