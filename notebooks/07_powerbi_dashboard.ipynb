# 07 - Power BI Dashboard Design

## Objective

Design an HR Attrition Intelligence Dashboard that converts machine learning predictions into actionable retention insights.

## Audience

- HR Managers
- HR Directors
- Business Leaders
- People Analytics Teams

## Dashboard Pages

### Page 1
Executive Summary

### Page 2
Department Analysis

### Page 3
Job Role Analysis

### Page 4
Employee Risk Explorer

# Page 1 - Executive Summary

Purpose:

Provide a high-level overview of organizational attrition risk.

## KPI Cards

1. Total Employees
2. High Risk Employees
3. Medium Risk Employees
4. Average Attrition Risk
5. Predicted Attrition Rate

## Visuals

Visual 1:
Risk Level Distribution (Donut Chart)

Visual 2:
Department-wise Average Risk (Bar Chart)

Visual 3:
Top 10 High Risk Employees (Table)

Visual 4:
Average Risk by Job Role (Bar Chart)

# Page 2 - Department Analysis

Purpose:

Identify departments requiring retention intervention.

## Visuals

1. Department vs Average Risk
2. Department vs High Risk Count
3. Department vs Monthly Income
4. Department Risk Heatmap

KPIs

- Highest Risk Department
- Lowest Risk Department

# Page 3 - Job Role Analysis

Purpose:

Identify job roles with elevated attrition risk.

## Visuals

1. Job Role vs Average Risk
2. Job Role vs High Risk Count
3. Job Role Risk Ranking
4. Attrition Probability Distribution

# Page 4 - Employee Risk Explorer

Purpose:

Enable HR users to identify employees requiring proactive retention action.

## Employee Table

Columns

- Employee Number
- Department
- Job Role
- Age
- Monthly Income
- Years At Company
- Attrition Probability
- Risk Level

Filters

- Department
- Job Role
- Risk Level

# Business Insights

Key Findings

- Sales department exhibits highest attrition risk.
- Sales Representatives represent the most vulnerable employee segment.
- Approximately 21% of employees fall into the High Risk category.
- Attrition risk can be monitored proactively using probability scores.
